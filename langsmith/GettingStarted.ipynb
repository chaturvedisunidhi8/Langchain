```python
#chatprompt template  (to tell llm behave like this)
from langchain_core.prompts import ChatPromptTemplate

prompt=ChatPromptTemplate.from_messages(
    [
    ("system","you are an  expert ai engineer"),
    ("user","{input}")
])
prompt
```

```python
chain=prompt|llm  ##means combining prompt with llm (chain helps to combine)
response=chain.invoke({"input":"can you tell me something about langsmith?"}) ##from here input get combine with prompt
#and given to llm
response
```

 from_template

You give just one text prompt with placeholders.

Example:

ChatPromptTemplate.from_template("Translate this to French: {text}")


 Its like filling in a blank in a sentence.

 from_messages

You give a list of messages with roles (system, user, assistant).

Example:

ChatPromptTemplate.from_messages([
    ("system", "You are a helpful translator."),
    ("user", "Translate this to French: {text}")
])


 Its like writing a mini chat script between you and the AI.
